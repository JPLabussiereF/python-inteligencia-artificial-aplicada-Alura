````python
import time

def resumir_email(corpos_emails):
  lista_de_resumos = []
  for index, email in enumerate(corpos_emails):
    email_resumido = client.models.generate_content(model='gemini-2.5-flash',
                                                    contents=f"""Vou te mandar o corpo de um email. Quero que voce resuma em apenas uma linha, passando o intuito desse email.
                                                    Segue o email: {email}.""")
    print(f'E-mail {index+1}: {email_resumido.text}')
    print("\n")

    lista_de_resumos.append(f'E-mail {index + 1}: {email_resumido.text}')

    time.sleep(6) # Colocado para não bater o rate-limite da API do Gemini
  return lista_de_resumos
````

Desafio:
1. Carregar um arquivo .csv com feedback de clientes (reviews.csv)
2. Usar um LLM para classificar o sentimento de cada feedback (Positivo, Negativo, Neutro)
3. Adicionar uma nova coluna com essa classificação ao DataFrame.

````python
# 1. Carregar um arquivo .csv com feedback de clientes (reviews.csv)
df_reviews = pd.read_csv('/content/reviews.csv')
df_reviews.head()
````

````python
coluna_reviews = df_reviews['reviewText']
````

### Few-shot Prompting

````python

# 2. Usar um LLM para classificar o sentimento de cada feedback (Positivo, Negativo, Neutro)


lista_feedbacks = []
for index, review in enumerate(coluna_reviews):
  feedbacks = client_groq.chat.completions.create(model='openai/gpt-oss-20b',
                                                 messages=[
                                                  {
                                                    "role": "user",
                                                    "content": f"""
Você é um classificador de sentimentos.
Sua tarefa é ler um texto de review e classificar apenas como 'Positiva', 'Negativa' ou 'Neutro'.
Não explique, não justifique, não adicione nada além da classificação.

Exemplos:

Review: 'O atendimento foi excelente, voltarei mais vezes!'
Classificação: Positiva

Review: 'A entrega demorou muito e o produto veio com defeito.'
Classificação: Negativa

Review: 'Produto razoável, nada demais, apenas ok.'
Classificação: Neutro

Agora siga a mesma lógica para o texto abaixo:

Lembre-se: Eu quero apenas as palavras 'Positiva', 'Negativa' ou 'Neutro'.

Review: {review}
                                                    """
                                                    }
                                                  ],
                                                    temperature=1,
                                                    max_completion_tokens=8192,
                                                    top_p=1,
                                                    reasoning_effort="medium",
                                                    stream=False,
                                                    stop=None
                                                )

  texto_feedbacks = feedbacks.choices[0].message.content
  lista_feedbacks.append(texto_feedbacks)
````

````python
lista_feedbacks
````

````python
# 3. Adicionar uma nova coluna com essa classificação ao DataFrame.

df_reviews['Análise de Sentimentos'] = lista_feedbacks
````
